```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Dataset
df=pd.read_csv(food_orderss.csv)

print("========== FOOD DELIVERY SALES ANALYSIS ==========")

# Task 1 - Analyze Order Data
print("\nOrder Data:")
print(df)

# Total Orders
print("\nTotal Orders:", len(df))

# Total Revenue
print("Total Revenue: Rs.", df['Amount'].sum())

# Task 2 - Peak Order Times
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M').dt.hour

peak_hours = df['Hour'].value_counts().sort_index()

print("\nPeak Order Hours:")
print(peak_hours)

# Task 3 - Most Popular Dishes
popular_items = df['Item'].value_counts()

print("\nMost Popular Dishes:")
print(popular_items)

# Location Analysis
location_orders = df['Location'].value_counts()

print("\nOrders by Location:")
print(location_orders)

# Dashboard

# Popular Dishes Chart
plt.figure(figsize=(6,4))
popular_items.plot(kind='bar')
plt.title("Most Popular Dishes")
plt.xlabel("Food Item")
plt.ylabel("Number of Orders")
plt.show()

# Peak Order Hours Chart
plt.figure(figsize=(6,4))
peak_hours.plot(kind='bar')
plt.title("Peak Order Hours")
plt.xlabel("Hour")
plt.ylabel("Orders")
plt.show()

# Location Pie Chart
plt.figure(figsize=(6,4))
location_orders.plot(kind='pie', autopct='%1.1f%%')
plt.title("Orders by Location")
plt.ylabel("")
plt.show()

print("\n========== BUSINESS INSIGHTS ==========")
print("Most Ordered Food Item:", popular_items.idxmax())
print("Peak Order Hour:", peak_hours.idxmax())
print("Highest Order Location:", location_orders.idxmax())
```
